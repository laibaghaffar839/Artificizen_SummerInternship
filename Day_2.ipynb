### Question # 1 
#### Use a list comprehension to extract all even numbers from a list in one line.

```python
list = [1, 23,67,88,66,55,44,42,33,22,11]
evens = [i for i in list if i % 2 == 0]
print(evens)
```

### Question # 2
#### Count the frequency of each word in a sentence using a dictionary.

```python
words = "Python is very good language and Python is very easy language."

word_list = words.split()

word_count = {}
for word in word_list:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1    

for word, count in word_count.items():
    print(f"{word}: {count}")
```

### Question # 3
#### Find duplicate elements in a list using a set.

```python
numbers = [10,20,40,25,32,10,23,32,20,40,20,56,78,56,89,32]

seen = set()
duplicates = set()
for number in numbers:
    if number in seen:
        duplicates.add(number)
    else:
        seen.add(number)

print("Duplicates:", duplicates)    
```

### Question # 4
#### Merge two dictionaries; if a key exists in both, sum the values

```python
dict1 = {'a': 167, 'b': 289, 'c': 3}
dict2 = {'b': 434, 'c': 500, 'd': 699}

for key in dict2:
    if key in dict1:
        dict1[key] = dict1[key] + dict2[key]
    else:
        dict1[key] = dict2[key]

print(f"Merged dictionary: {dict1}")

```

### Question # 5
#### Given a sentence, count the vowels and consonants separately

```python
def vow_con_count(sentence):
    vowel_count = 0
    consonant_count = 0
    vowels = "aeiouAEIOU"
    
    for char in sentence:
        if char.isalpha():   # Only count letters
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1
    return vowel_count, consonant_count

Sentence = "Python is very good language and Python is very easy language."
a, b = vow_con_count(Sentence)
print(f"Vowels: {a}\nConsonants: {b}")
```

### Question # 6
#### Write a function that returns the most frequent element in a list.

```python
def freq_element(lst):
    return max(set(lst), key=lst.count)


list = [10,11,10,20,10,11,78,32,78]
element = freq_element(list)
print(f"Most frequent element: {element}")
```
