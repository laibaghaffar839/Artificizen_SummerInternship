### Question # 1
#### Create a base class Animal with a speak() method, and Dog/Cat subclasses that override it.

```python
class Animal:
    def speak(self):
        print("The animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("The dog barks")

class Cat(Animal):
    def speak(self):
        print("The cat meows")

animal = Animal()
animal.speak()  
dog = Dog()
dog.speak()  
cat = Cat()
cat.speak()  #Animal is the base class with a speak() method. Dog and Cat inherit from Animal.Both subclasses override the speak() method to provide their own behavior.
```

### Question # 2
#### Define a custom exception InsufficientBalanceError and raise it inside the BankAccount class from Day 4.

```python
class BankAccount:
    def __init__(self, acc_no, acc_holder, balance):
        self.acc_no = acc_no
        self.acc_holder = acc_holder
        self.balance = balance
    
    def deposit(self, amount):
        if amount <= 0:
            print("Deposit amount must be positive")
            return
        self.balance += amount
        print(f"Deposited: ${amount}. New Balance: ${self.balance}")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive")
            return
        if amount > self.balance:
            raise InsufficientBalanceError("Insufficient Balance!")
        else:
            self.balance -= amount
            print(f"Withdrew: ${amount}. New Balance: ${self.balance}")

class InsufficientBalanceError(Exception):
    pass
        
print("\nBank Account Information:\n")
print("Account 1:")
acc1 = BankAccount("123456", "Ahmed", 100000)
try:
    acc1.deposit(5000)
    acc1.withdraw(2000)
    acc1.withdraw(150000)  # Attempt to withdraw more than balance
except InsufficientBalanceError as In:
    print(In)


print("\nAccount 2:")
acc2 = BankAccount("654321", "Ali", 50000)
acc2.deposit(10000)
acc2.withdraw(30000)
acc2.withdraw(-5000)  # Attempt to withdraw negative amount
```

### Question # 3
#### Write a program that reads a text file and reports the number of lines, words, and characters.

```python
with open("data.txt",'w') as file:
    file.write("Hello, this is a sample text file\nPython is great for file handling!\nProgramming is fun!")

with open("data.txt",'r') as file:
    content = file.read()
    print("\nFile Content:\n")
    print(content)

lines = len(content.splitlines())
print(f"\nNumber of lines in the file: {lines}")

words = len(content.split())
print(f"\nNumber of words in the file: {words}")

character = len(content)
print(f"\nNumber of characters in the file content: {character} characters")



```

### Question # 4
#### Write a program that saves a list of dictionaries (student records) to a JSON file, then loads them back.

```python
import json

student_data = [
    {
        "name": "Ahmed",
        "age": 20,
        "major": "Computer Science"
    },
    {
        "name": "Ali",
        "age": 22,
        "major": "Mathematics"
    },
    {
        "name": "Sara",
        "age": 21,
        "major": "Physics"
    }
]

with open("students.json", "w") as json_file:
    json.dump(student_data, json_file, indent=4)

with open("students.json", "r") as json_file:
    data = json.load(json_file)
print("\nStudent Data from JSON:\n")
print(data)
print(data[0])  
print(data[1]["name"])

```

### Question # 5
#### Implement __str__ and __repr__ for a Point(x, y) class so printing an instance gives readable output.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Point({self.x}, {self.y})"
    
    def __repr__(self):
        return f"Point at (x={self.x}, y={self.y})"
    
p = Point(30, 40)

print("\nString Representation of Point:\n") # user friendly
print(p)     # also uses __str__
print(str(p))   
print("\nRepr Representation of Point:\n")
print(repr(p))   # developer-friendly
```
