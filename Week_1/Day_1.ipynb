### Question No.1 Write a program to check whether a number is prime.

```python
num = int(input("Enter Number:"))

if num <= 1:
    print(num, "is not a prime number")
else:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
```

### Question No. 2 Reverse a string without using slicing or a built-in reverse function.

```python
text = input("Enter a string: ")

reverse = ""

for char in text:
    reverse = char + reverse

print(reverse)
```

### Question No.3 Print the Fibonacci series up to n terms.

```python
n = int(input("Enter Number:"))

a = 0
b = 1
c = 0
while (c <= n):
    print(c)
    a = b
    b = c
    c = a + b
```

### Question No.4 FizzBuzz: print numbers 1 to 100, replacing multiples of 3 with “Fizz”, multiples of
5 with “Buzz”, and multiples of both with “FizzBuzz”.

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

### Question No.5 Check if a given string is a palindrome.

```python
text=input()

reverse=""

for char in text:
    reverse=char+reverse

if text==reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

### Question No.6 Given a list of numbers, find the largest and smallest without using min()/max()

```python
num = [12, 45, 7, 89, 23, 56]

sorted_num = sorted(num)

smallest = sorted_num[0]
largest = sorted_num[-1]

print("Smallest:", smallest)
print("Largest:", largest)
```
