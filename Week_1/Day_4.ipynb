### Question # 1
#### Create a Car class with attributes (brand, model, year) and a method that displays the car’s info.

```python
class Car:
    def __init__(self,brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
    def display_info(self):
        print(f"Car Brand: {self.brand}")
        print(f"Car Model: {self.model}")
        print(f"Car Year: {self.year}")

print("Cars Information:\n")

print("Car 1:")
car1 = Car("Toyota", "Camry", 2020)
car1.display_info()
print("\n")

print("Car 2:")
car2= Car("Honda", "Civic", 2019)
car2.display_info()
```

### Question # 2
#### Create a BankAccount class with deposit() and withdraw() methods; prevent withdrawals beyond the balance.

```python
class BankAccount:
    def __init__(self, acc_no, acc_holder, balance):
        self.acc_no = acc_no
        self.acc_holder = acc_holder
        self.balance = balance
    
    def deposit(self, amount):
        if amount <= 0:
            print("Deposit amount must be positive")
            return
        self.balance += amount
        print(f"Deposited: ${amount}. New Balance: ${self.balance}")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive")
            return
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"Withdrew: ${amount}. New Balance: ${self.balance}")
        
print("\nBank Account Information:\n")
print("Account 1:")
acc1 = BankAccount("123456", "Ahmed", 100000)
acc1.deposit(5000)
acc1.withdraw(2000)
acc1.withdraw(150000)  # Attempt to withdraw more than balance

print("\nAccount 2:")
acc2 = BankAccount("654321", "Ali", 50000)
acc2.deposit(10000)
acc2.withdraw(30000)
acc2.withdraw(-5000)  # Attempt to withdraw negative amount
```

### Question # 3
#### Demonstrate the difference between instance and class variables: track how many objects of a class have been created.

```python
class Cars():

    total_cars = 0        #Class variable
    Country =  "USA"      #Class variable

    def __init__(self, name, model, year):
        self.name = name     #Instance variable
        self.model = model   #Instance variable
        self.year = year     #Instance variable
        Cars.total_cars += 1

car1 = Cars("Toyota", "Camry", 2020)

car2 = Cars("Honda", "Civic", 2019)

car3 = Cars("BMW", "X5", 2021)

print("\nTotal Cars:", Cars.total_cars)
print("Country:", Cars.Country)
```

An instance variable belongs to an individual object, so each object has its own separate value. It is created using self inside the __init__() method. A class variable belongs to the class and is shared by all objects. It is declared directly inside the class, outside any method. Class variables are used to store information common to all objects, such as counting the total number of objects created

### Question # 4
#### Write a class with a private attribute, exposing it safely through @property

```python
class BankAccount:

    def __init__(self, balance):
        # private attribute (hidden using name mangling)
        self.__balance = balance

    # getter (safe access)
    @property
    def balance(self):
        print("Accessing balance safely...")
        return self.__balance

    # setter (controlled update)
    @balance.setter
    def balance(self, value):
        if value >= 0:
            print("Updating balance safely...")
            self.__balance = value
        else:
            print("Error: Balance cannot be negative!")

account = BankAccount(1000)


print(account.balance)     # safe read (getter)


account.balance = 2000     # safe update (setter)
print(account.balance)     # safe read (getter)

account.balance = -69    # attempt to set a negative balance (setter will prevent this)

  
```

### Question # 5
#### Add a @classmethod to a Person class that works as an alternate constructor, e.g. Person.from_birth_year(name, year).

```python
from datetime import date

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def from_birth_year(cls, name, year):
        current_year = date.today().year
        age = current_year - year
        return cls(name, age)
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")
    
# Creating object using alternate constructor
p1 = Person.from_birth_year("Ali", 2002)

p1.display_info()

p2 = Person("Ahmed", 20)
p2.display_info()
```
