### Question # 1
#### Write a function that accepts any number of arguments and returns their sum
(use *args).

```python
def number(*args):
    return sum(args)
num = number(23,45,1,3)

print(num)
```

```python
def show(**kwargs):
    print(kwargs)
show(name="Alice", age=30)
```

### Question # 2
#### Use map() and a lambda to square every number in a list.

```python
num = [2,4,5,6,7,8,9]

square = list(map(lambda x: x**2, num))

print(square)
```

### Question # 3
#### Use filter() to extract all strings longer than 5 characters from a list.

```python
names = ["Lahore", "Karachi", "Islamabad", "Peshawar", "Pakistan", "Nepal", "India", "China", "Bangladesh"]

long_names = list(filter(lambda x:len(x)>5,names))
print(long_names)
```

### Question # 4
#### Write a recursive function to calculate the factorial of a number.

```python
# With loop
n = int(input("Enter a number: "))
fact = 1

for i in range(1,n+1):
    fact *= i

print("Factorial of", n, "is", fact)
```

```python
# with recursion
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

n = int(input("Enter a number: "))
print("Factorial of", n, "is", factorial(n))
```

### Question # 5
#### Sort a list of dictionaries (e.g., students with name and marks) by marks using
sorted() and a key function.

```python
students = [
    {"name":"Dua", "marks": 90},
    {"name":"Ali", "marks": 80},
    {"name":"Sara", "marks": 85},
    {"name":"Ahmed", "marks": 95},
    {"name":"Ayesha", "marks": 75}
    ]
students_marks = list(sorted(students, key=lambda x:x['marks']))
print(students_marks)
```

### Question # 6
#### Write a function with a default argument, and add a comment explaining why
mutable default arguments (like a list) are risky in Python.

```python
def add_items(item,my_list = None): # Using None is safer than using a list ([]) as a default argument.
    if my_list is None:  
        my_list = []        # Mutable default arguments are risky because the same list is reused
                            # every time the function is called, which can cause unexpected results.
    my_list.append(item)
    return my_list

list1 = add_items(1)
list2 = add_items(2)    
list3 = add_items(3, [4,5,6])
print(list1, list2, list3, sep="\n")
```

Mutable default arguments (like [] or {} , set()) are risky because they are created only once when the function is defined. If the function modifies them, the changes remain for future function calls. To avoid this, use None as the default value and create a new list inside the function.
